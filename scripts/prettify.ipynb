```python
import sqlite3
import json

from pathlib import Path
```

```python
con = sqlite3.connect("predictions.db")
con.row_factory = sqlite3.Row
cur = con.cursor()

base = Path("./json")

if not base.exists():
    base.mkdir()
```

```python
sql = """
SELECT
    ida as a_id,
    sources.publication as a_publication,
    sources.date as a_date,
    sources.author as a_author,
    sources.url as a_url,
    sources.text as a_text,
    idb as b_id,
    texts.publication as b_publication,
    texts.date as b_date,
    texts.author as b_author,
    texts.url as b_url,
    texts.text as b_text,
    scores
FROM
    predictions 
INNER JOIN
    sources ON ida=sources.id 
INNER JOIN
    texts ON idb=texts.id 
WHERE
    label=1
"""

rowiter = cur.execute(sql)
```

```python
for n, row in enumerate(rowiter):
    data = {key: row[key] for key in row.keys()}
    
    # if not len(data["a_text"]) < 2000 and not len(data["b_text"]) < 2000:
    with (base / f"{n}.json").open(mode="w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
```
