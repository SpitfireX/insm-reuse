```python
import sqlite3

from pathlib import Path
```

```python
destcon = sqlite3.connect("predictions.db")
destcur = destcon.cursor()
destcur.execute("""
CREATE TABLE IF NOT EXISTS predictions (
    ida INTEGER NOT NULL,
    idb INTEGER NOT NULL,
    scores TEXT,
    label TEXT,
    FOREIGN KEY(ida) REFERENCES sources(id)
    FOREIGN KEY(idb) REFERENCES texts(id)
    PRIMARY KEY (ida, idb)
);
""")

base = Path(".")
```

```python
for p in base.glob("clue*.db"):
    srccon = sqlite3.connect(p)
    srccur = srccon.cursor()

    rowiter = srccur.execute("SELECT * FROM predictions;")

    for row in rowiter:
        destcur.execute("INSERT OR IGNORE INTO predictions VALUES(?, ?, ?, ?)", row);
    
    destcon.commit()
    
    srccon.close()
        
destcon.close()
```
